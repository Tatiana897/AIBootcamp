Later the Street column was cleaned with the Street in front of each data

```python
# Функция для стандартизации названий улиц
def standardize_address_street(street):
    if pd.isna(street) or street.strip().lower() == 'nan':
        return np.nan  # Заменить пустые значения на NaN
    
    # Словарь замен для стандартизации
    replacements = {
        r"\bst\.?\b": "Street",  # "st", "st." -> "Street"
        r"\bstraße\b": "Street",  # "straße" -> "Street"
        r"\bchemin\b": "Street",  # "chemin" -> "Street"
        r"\brue\b": "Street",  # "rue" -> "Street"
        r"\bincrocio\b": "Intersection",  # "incrocio" -> "Intersection"
        r"\bglorieta\b": "Circle",  # "glorieta" -> "Circle"
    }
    
    # Привести строку к нижнему регистру и удалить пробелы
    street = street.strip().lower()
    
    # Заменить синонимы с помощью словаря (регулярные выражения)
    for pattern, replacement in replacements.items():
        street = re.sub(pattern, replacement, street)
    
    # Разбить строку на слова и капитализировать каждое
    words = street.split()
    words = [word.capitalize() for word in words]
    
    # Объединить слова обратно в строку
    street = " ".join(words)
    
    return street

# Применяем функцию к колонке address.street
df['address.street'] = df['address.street'].apply(standardize_address_street)
df
```

```python
def process_related_id(value):
    # Skip values that are not strings (e.g., NaN)
    if not isinstance(value, str):
        return value

    # Replace hyphen or period after 'CUST' and 'TRX' with an underscore
    value = value.replace('CUST-', 'CUST_').replace('CUST.', 'CUST_')
    value = value.replace('TRX-', 'TRX_').replace('TRX.', 'TRX_')

    # If no delimiter after 'CUST' or 'TRX', add an underscore
    if value.startswith("CUST") and not value[4] == "_":
        value = "CUST_" + value[4:]
    elif value.startswith("TRX") and not value[3] == "_":
        value = "TRX_" + value[3:]

    # If CUST or TRX is at the end, move them to the front and add an underscore
    if value.endswith("CUST") or value.endswith("TRX"):
        if value.endswith("CUST"):
            value = "CUST_" + value.replace("CUST", "").strip("_")
        if value.endswith("TRX"):
            value = "TRX_" + value.replace("TRX", "").strip("_")

    # Remove any redundant or extra underscores
    value = "_".join([part for part in value.split("_") if part])

    return value

# Apply to the 'related_id' column
lgs['related_id'] = lgs['related_id'].apply(process_related_id)
lgs.head()
```

```python
# Функция для обработки веса
def process_weight(specs):
    if 'weight' in specs:
        weight = specs['weight']
        if 'kg' in weight:
            return round(float(weight.replace('kg', '').strip()), 2)
        elif 'lbs' in weight:
            return round(float(weight.replace('lbs', '').strip()) * 0.453592, 2)  # Конвертация в кг
    return None

# Функция для обработки размеров
def process_dimensions(specs):
    if 'dimensions' in specs:
        dims = specs['dimensions']
        dims = re.findall(r"[\d.]+", dims)  # Извлекаем числа
        return [round(float(d), 2) for d in dims]
    return None

# Применяем функции и добавляем новые колонки
prod['weight_kg'] = prod['specifications'].apply(process_weight)
prod['dimensions'] = prod['specifications'].apply(process_dimensions)

# Удаляем колонку specifications
prod = prod.drop(columns=['specifications'])

# Перемещаем новые колонки рядом друг с другом
columns = list(prod.columns)
columns.insert(columns.index('price'), columns.pop(columns.index('weight_kg')))
columns.insert(columns.index('price'), columns.pop(columns.index('dimensions')))
prod = prod[columns]

prod
```

Later price column was brought to one form by dividing into two columns price_numeric and currency

```python
# Функция для выделения числовой части и валюты
def process_price(value):
    import re
    # Регулярное выражение для обработки чисел и валют в любом порядке
    match = re.match(r"([€$¥]?\s*[\d,.k]+)\s*(\w+|[$€¥])?", value)
    if match:
        price_part = match.group(1).replace("k", "000").replace(",", "")
        currency_part = match.group(2)
        
        # Если валюта не найдена, то проверим, был ли символ валюты в начале
        if not currency_part:
            if price_part.startswith("€"):
                currency_part = "EUR"
                price_part = price_part[1:].strip()  # Убираем евро-символ
            elif price_part.startswith("$"):
                currency_part = "USD"
                price_part = price_part[1:].strip()  # Убираем долларовый символ
            elif price_part.startswith("¥"):
                currency_part = "JPY"
                price_part = price_part[1:].strip()  # Убираем йеновый символ
        
        # Преобразование валютных символов в текстовый формат
        price = float(price_part)
        currency = currency_part.replace("$", "USD").replace("€", "EUR").replace("¥", "JPY")
        
        return price, currency
    return np.nan, np.nan  

# Создаем новые колонки: price_numeric и currency
prod[['price_numeric', 'currency']] = prod['price'].apply(lambda x: pd.Series(process_price(x)))

# Удаляем оригинальную колонку price
prod = prod.drop(columns=['price'])

# Перемещаем новые колонки рядом с 'stock_level'
columns = list(prod.columns)
columns.insert(columns.index('stock_level'), columns.pop(columns.index('price_numeric')))
columns.insert(columns.index('stock_level'), columns.pop(columns.index('currency')))
prod = prod[columns]

prod
```

```python
import pandas as pd
import re

# Функция для преобразования десятичных координат в DMS с учетом знака
def decimal_to_dms(decimal, is_latitude=True):
    # Определяем знак
    degrees = int(decimal)
    minutes = int((decimal - degrees) * 60)
    seconds = round(((decimal - degrees) * 60 - minutes) * 60, 1)
    
    # Если координаты для широты
    if is_latitude:
        direction = 'N' if decimal >= 0 else 'S'  # Для широты 'N' или 'S'
    else:
        direction = 'E' if decimal >= 0 else 'W'  # Для долготы 'E' или 'W'
    
    return f"{abs(degrees)}°{abs(minutes)}'{abs(seconds)}\"{direction}"

# Функция для обработки координат
def process_coordinates(value):
    if isinstance(value, str):
        # Проверяем, если уже в формате DMS
        match = re.match(r"(\d+)°(\d+)'([\d\.]+)\"([NS])\s*,\s*(\d+)°(\d+)'([\d\.]+)\"([EW])", value.strip())
        if match:
            # Если координаты уже в формате DMS, возвращаем их без изменений
            return value.strip()
        
        # Если координаты в десятичном формате, преобразуем их в DMS
        match = re.match(r"([\-\d\.\+]+)\s*,\s*([\-\d\.\+]+)", value.strip())
        if match:
            lat, lon = match.groups()
            lat_decimal = float(lat)
            lon_decimal = float(lon)
            lat_dms = decimal_to_dms(lat_decimal, is_latitude=True)
            lon_dms = decimal_to_dms(lon_decimal, is_latitude=False)
            return f"{lat_dms}, {lon_dms}"
    
    return value  # Если координаты не распарсены, возвращаем исходное значение

# Применяем функцию к колонке 'location.coordinates'
stores['location.coordinates'] = stores['location.coordinates'].apply(process_coordinates)

stores
```

```python
# Adjusted function to handle multiple currencies and formats
def vectorized_process_price(amount_series):
    prices = []
    currencies = []
    
    # Enhanced regex to capture various currency symbols and formats (including "K" for thousands)
    currency_pattern = r'([^\d]+)?([\d,]+(?:\.\d{1,2})?)([a-zA-Z]+)?'

    for value in amount_series:
        match = re.match(currency_pattern, value)
        if match:
            currency = match.group(1) or match.group(3) or 'USD'  # Get currency from symbol or code, default to 'USD'
            price = match.group(2).replace(',', '')  # Remove commas from price

            # Handle 'K' format (e.g., 5.5K)
            if 'K' in price.upper():
                price_numeric = float(price.replace('K', '').strip()) * 1000
            else:
                try:
                    price_numeric = float(price)
                except ValueError:
                    price_numeric = None  # Handle invalid price
        else:
            currency = 'USD'  # Default currency
            price_numeric = None  # Handle missing or invalid price
            
        prices.append(price_numeric)
        currencies.append(currency)
    
    return prices, currencies

# Apply to the 'total_amount' column
prices, currencies = vectorized_process_price(tran['total_amount'])
tran['price_numeric'] = prices
tran['currency'] = currencies

# Drop the original 'total_amount' column if necessary
tran = tran.drop(columns=['total_amount'])
tran
```
